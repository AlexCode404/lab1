#Задача 1
Написать функцию на вход которой подается строка, состоящая из латинских букв.
Функция должна вернуть количество гласных букв (a, e, i, o, u) в этой строке. Написать тесты для кода

```python
def get_vowel_count(input_string):
    """
    Подсчитывает количество гласных букв в строке.

    Args:
        input_string (str): Входная строка для анализа

    Returns:
        int: Количество гласных букв в строке
    """
    vowels = 'aeiouAEIOU'
    return sum(1 for char in input_string if char in vowels)

def test_get_vowel_count():
    assert get_vowel_count("hello") == 2
    assert get_vowel_count("HELLO") == 2
    assert get_vowel_count("aeiou") == 5
    assert get_vowel_count("bcdfg") == 0
    assert get_vowel_count("") == 0
    assert get_vowel_count("Python") == 1
    print("Тесты для get_vowel_count пройдены")

```

# Задача 2
Написать функцию на вход, которой подается строка. Функция должна вернуть true, если
каждый символ в строке встречается только 1 раз, иначе должна вернуть false. Написать тесты для кода

```python
def unique_chars(input_string):
    """
    Проверяет, состоит ли строка только из уникальных символов.

    Args:
        input_string (str): Строка для проверки

    Returns:
        bool: True, если все символы уникальны, иначе False
    """
    return len(input_string) == len(set(input_string))


def test_unique_chars():
    assert unique_chars("abcdef") == True
    assert unique_chars("hello") == False
    assert unique_chars("") == True
    assert unique_chars("a") == True
    assert unique_chars("aA") == True
    assert unique_chars("abcdefg") == True
    assert unique_chars("programming") == False
    print("Тесты для unique_chars пройдены")

```

# Задача 3
Написать функцию, которая принимает положительное число и возвращает количество
бит равных 1 в этом числе. Написать тесты для кода

```python
def count_bits(number):
    """
    Подсчитывает количество единичных битов в двоичном представлении числа.

    Args:
        number (int): Целое число для анализа

    Returns:
        int: Количество единичных битов в двоичном представлении
    """
    return bin(number).count('1')


def test_count_bits():
    assert count_bits(0) == 0
    assert count_bits(1) == 1
    assert count_bits(2) == 1
    assert count_bits(7) == 3
    assert count_bits(15) == 4
    assert count_bits(8) == 1
    assert count_bits(255) == 8
    print("Тесты для count_bits пройдены")

```

# Задача 4
Написать функцию, которая принимает положительное число. Функция должна вернуть то,
сколько раз необходимо перемножать цифры числа или результат перемножения, чтобы
получилось число состоящее из одной цифры.
Например, для входного числа:
· 39 функция должна вернуть 3, так как 3*9=27 => 2*7=14 => 1*4=4
· 4 функция должна вернуть 0, так как число уже состоит из одной цифры
· 999 функция должна вернуть 4, так как 9*9*9=729 => 7*2*9=126 => 1*2*6=12 =>
1*2=2. Написать тесты для кода

```python
def count_multiplication_steps(number):
    """
    Вычисляет количество итераций до однозначного числа.

    Args:
        number (int): Неотрицательное целое число

    Returns:
        int: Количество итераций (персистентность числа)
    """
    count = 0
    while number >= 10:
        product = 1
        temp = number
        while temp > 0:
            product *= temp % 10
            temp //= 10
        number = product
        count += 1
    return count

def test_count_multiplication_steps():
    assert count_multiplication_steps(39) == 3
    assert count_multiplication_steps(4) == 0
    assert count_multiplication_steps(999) == 4
    assert count_multiplication_steps(10) == 1
    assert count_multiplication_steps(25) == 2
    assert count_multiplication_steps(123) == 1
    print("Тесты для count_multiplication_steps пройдены")

```

# Задача 5
Написать функцию, которая принимает два целочисленных вектора одинаковой длины и
возвращает среднеквадратическое отклонение двух векторов. Написать тесты для кода

```python
def calculate_mse(predicted_values, actual_values):
    """
    Вычисляет среднеквадратическое отклонение между двумя векторами.

    Args:
        predicted_values (list): Список предсказанных значений
        actual_values (list): Список истинных значений

    Returns:
        float: Значение среднеквадратического отклонения

    Raises:
        ValueError: Если векторы имеют разную длину
    """
    if len(predicted_values) != len(actual_values):
        raise ValueError("Векторы должны быть одинаковой длины")
    if len(predicted_values) == 0:
        return 0.0
    squared_errors = [(p - t) ** 2 for p, t in zip(predicted_values, actual_values)]
    return sum(squared_errors) / len(squared_errors)


def test_calculate_mse():
    assert calculate_mse([1, 2, 3], [1, 2, 3]) == 0
    assert calculate_mse([1, 2, 3], [2, 3, 4]) == 1
    assert calculate_mse([0, 0, 0], [1, 1, 1]) == 1
    assert abs(calculate_mse([1.5, 2.5], [1, 2]) - 0.25) < 0.0001
    assert calculate_mse([10], [5]) == 25
    assert calculate_mse([], []) == 0.0
    assert calculate_mse([5], [3]) == 4.0
    assert calculate_mse([-1, -2], [1, 2]) == 10.0
    print("Тесты для calculate_mse пройдены")
```

# Задача 6
Написать функцию, принимающая целое положительное число. Функция должна вернуть
строку вида “(n1**p1)(n2**p2)…(nk**pk)” представляющая разложение числа на простые
множители (если pi == 1, то выводить только ni).
Например, для числа 86240 функция должна вернуть “(2**5)(5)(7**2)(11)”. Написать тесты для кода

```python
def prime_factorization(number):
    """
    Выполняет разложение числа на простые множители.

    Args:
        number (int): Положительное целое число для разложения

    Returns:
        str: Строка с простыми множителями в формате "(p1**k1)(p2**k2)..."

    Raises:
        ValueError: Если число меньше 2
    """
    if number < 2:
        return ""

    factors = {}
    d = 2
    while d * d <= number:
        while number % d == 0:
            factors[d] = factors.get(d, 0) + 1
            number //= d
        d += 1
    if number > 1:
        factors[number] = factors.get(number, 0) + 1

    result = []
    for prime in sorted(factors.keys()):
        power = factors[prime]
        if power == 1:
            result.append(f"({prime})")
        else:
            result.append(f"({prime}**{power})")

    return "".join(result)


def test_prime_factorization():
    assert prime_factorization(86240) == "(2**5)(5)(7**2)(11)"
    assert prime_factorization(2) == "(2)"
    assert prime_factorization(12) == "(2**2)(3)"
    assert prime_factorization(17) == "(17)"
    assert prime_factorization(100) == "(2**2)(5**2)"
    assert prime_factorization(1) == ""
    print("Тесты для prime_factorization пройдены")
```

# Задача 7
Написать функцию, принимающая целое число n, задающее количество кубиков. Функция
должна определить, можно ли из данного кол-ва кубиков построить пирамиду, то есть
можно ли представить число n как 1^2+2^2+3^2+…+k^2. Если можно, то функция должна
вернуть k, иначе строку “It is impossible”. Написать тесты для кода

```python
def pyramid(number):
    """
    Определяет, можно ли построить пирамиду из заданного количества кубиков.

    Args:
        number (int): Количество доступных кубиков

    Returns:
        int or str: Высота пирамиды, если построение возможно,
                   иначе "It is impossible"
    """
    if number < 1:
        return "It is impossible"

    k = 1
    total = 0
    while total < number:
        total += k * k
        if total == number:
            return k
        k += 1

    return "It is impossible"


def test_pyramid():
    assert pyramid(1) == 1
    assert pyramid(5) == 2
    assert pyramid(14) == 3
    assert pyramid(30) == 4
    assert pyramid(55) == 5
    assert pyramid(6) == "It is impossible"
    assert pyramid(0) == "It is impossible"
    print("Тесты для pyramid пройдены")

```

# Задача 8
Функция принимает на вход положительное число и определяет является ли оно сбалансированным, т.е. сумма цифр до средних равна сумме цифр после. Средними в случае нечетного числа цифр считать одну цифру, в случае четного - две средних. Написать тесты для кода

```python
def is_balanced(n):
    """
    Проверяет, является ли число сбалансированным.

    Args:
        n (int): Положительное целое число для проверки

    Returns:
        bool: True, если число сбалансированное, иначе False
    """
    s = str(n)
    length = len(s)

    if length <= 2:
        return True

    if length % 2 == 1:
        # Нечетное количество цифр - одна средняя
        mid = length // 2
        left_sum = sum(int(s[i]) for i in range(mid))
        right_sum = sum(int(s[i]) for i in range(mid + 1, length))
    else:
        # Четное количество цифр - две средние
        mid = length // 2
        left_sum = sum(int(s[i]) for i in range(mid - 1))
        right_sum = sum(int(s[i]) for i in range(mid + 1, length))

    return left_sum == right_sum


def test_is_balanced():
    assert is_balanced(1) == True
    assert is_balanced(12) == True
    assert is_balanced(1234) == False
    assert is_balanced(12321) == True
    assert is_balanced(1234321) == True
    assert is_balanced(56239814) == True
    assert is_balanced(9999) == True
    print("Тесты для is_balanced пройдены")
```
